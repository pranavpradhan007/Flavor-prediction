```python
import pandas as pd
```

```python
receptor_search=pd.read_csv('Data\BitterDB\ReceptorSearchA_2024.csv')
receptor_search.head()
```

```python
receptor_search.isna().sum()
```

```python
receptor_search.dtypes
```

```python
relevant_columns = [
    'rID',          # Receptor ID
    'rName',        # Receptor Name
    'classification', # Classification of the receptor
    'organism',     # Organism information
    'MW',           # Molecular Weight
    'NumLigands',   # Number of ligands interacting with the receptor
    'proteinNames'  # Protein names associated with the receptor
]
receptor_search_cleaned = receptor_search[relevant_columns]
```

```python
# For MW, fill missing values with the median
receptor_search_cleaned['MW'] = receptor_search_cleaned['MW'].fillna(receptor_search_cleaned['MW'].median())

# For categorical columns, fill missing values with 'Unknown'
for col in ['rName', 'classification', 'organism', 'proteinNames']:
    receptor_search_cleaned[col] = receptor_search_cleaned[col].fillna('Unknown')
```

```python
receptor_search_cleaned.isna().sum()
```

```python
receptor_search_cleaned.to_csv("receptorSearchA_2024_cleaned.csv", index=False)
```
