```python
import pandas as pd
```

```python
# BitterDB

BitterCompoundsPropA=pd.read_csv('Data\BitterDB\Cleaned\BitterCompoundsPropA_2024_Cleaned.csv')
citterresourceA=pd.read_csv('Data\BitterDB\Cleaned\cbitterresourcesA_2024_final_cleaned.csv')
compoundsnamesA=pd.read_csv('Data\BitterDB\Cleaned\compoundsnamesA_2024_cleaned.csv')
ligandReceptorsA=pd.read_csv('Data\BitterDB\Cleaned\ligandReceptorsA_2024_cleaned.csv')
ReceptorSearchA=pd.read_csv('Data\BitterDB\Cleaned\\receptorSearchA_2024_cleaned.csv')

# FooDB

CompoundExternalDescriptor=pd.read_csv('Data\FooDB\\foodb_2020_4_7_csv\Cleaned\CompoundExternalDescriptor_cleaned.csv')
Compound=pd.read_csv('Data\FooDB\\foodb_2020_4_7_csv\Cleaned\compounds_cleaned.csv')
CompoundFlavor=pd.read_csv('Data\FooDB\\foodb_2020_4_7_csv\Cleaned\compoundsflavours_cleaned.csv')
CompoundSubstituent=pd.read_csv('Data\FooDB\\foodb_2020_4_7_csv\Cleaned\compoundsubstitute_cleaned.csv')
CompoundSynonym=pd.read_csv('Data\FooDB\\foodb_2020_4_7_csv\Cleaned\compoundsynonym_cleaned.csv')
Flavor=pd.read_csv('Data\FooDB\\foodb_2020_4_7_csv\Cleaned\\flavour_cleaned.csv')

# SweetnerDB

Sweetner=pd.read_csv('Data\SweetnerDB\SweetenersDB_v2.0.csv')
```

```python
import os
# import pandas as pd

def print_dataset_columns(dataset_name, file_path):
    """Prints columns of a dataset if file exists"""
    print(f"\n{dataset_name}:")
    if os.path.exists(file_path):
        try:
            df = pd.read_csv(file_path)
            print(f"Columns: {df.columns.tolist()}")
            print(f"Shape: {df.shape}")
        except Exception as e:
            print(f"Error reading file: {str(e)}")
    else:
        print("File not found")

# BitterDB datasets
print("=== BitterDB ===")
print_dataset_columns(
    "BitterCompoundsPropA",
    os.path.join("Data", "BitterDB", "Cleaned", "BitterCompoundsPropA_2024_Cleaned.csv")
)
print_dataset_columns(
    "citterresourceA",
    os.path.join("Data", "BitterDB", "Cleaned", "cbitterresourcesA_2024_final_cleaned.csv")
)
print_dataset_columns(
    "compoundsnamesA",
    os.path.join("Data", "BitterDB", "Cleaned", "compoundsnamesA_2024_cleaned.csv")
)
print_dataset_columns(
    "ligandReceptorsA",
    os.path.join("Data", "BitterDB", "Cleaned", "ligandReceptorsA_2024_cleaned.csv")
)
print_dataset_columns(
    "ReceptorSearchA",
    os.path.join("Data", "BitterDB", "Cleaned", "receptorSearchA_2024_cleaned.csv")
)

# FooDB datasets
print("\n=== FooDB ===")
print_dataset_columns(
    "CompoundExternalDescriptor",
    os.path.join("Data", "FooDB", "foodb_2020_4_7_csv", "Cleaned", "CompoundExternalDescriptor_cleaned.csv")
)
print_dataset_columns(
    "Compounds",
    os.path.join("Data", "FooDB", "foodb_2020_4_7_csv", "Cleaned", "compounds_cleaned.csv")
)
print_dataset_columns(
    "CompoundFlavor",
    os.path.join("Data", "FooDB", "foodb_2020_4_7_csv", "Cleaned", "compoundsflavours_cleaned.csv")
)
print_dataset_columns(
    "CompoundSubstituent",
    os.path.join("Data", "FooDB", "foodb_2020_4_7_csv", "Cleaned", "compoundsubstitute_cleaned.csv")
)
print_dataset_columns(
    "CompoundSynonym",
    os.path.join("Data", "FooDB", "foodb_2020_4_7_csv", "Cleaned", "compoundsynonym_cleaned.csv")
)
print_dataset_columns(
    "Flavors",
    os.path.join("Data", "FooDB", "foodb_2020_4_7_csv", "Cleaned", "flavour_cleaned.csv")
)

# SweetenerDB
print("\n=== SweetenerDB ===")
print_dataset_columns(
    "SweetenersDB",
    os.path.join("Data", "SweetnerDB", "SweetenersDB_v2.0.csv")
)
```

```python
# Standardize column names across datasets
BitterCompoundsPropA = BitterCompoundsPropA.rename(columns={'cid': 'compound_id'})
Compound = Compound.rename(columns={'id': 'compound_id'})
Sweetener = Sweetner.rename(columns={'ID': 'compound_id', 'Smiles': 'canonical_smiles'})
```

```python
# Select critical columns from each source
bitter_core = BitterCompoundsPropA[[
    'compound_id', 'canonical_smiles', 'MW', 'AlogP', 
    'Num_H_Acceptors', 'Num_H_Donors', 'Num_RotatableBonds'
]]

foodb_core = Compound[[
    'compound_id', 'moldb_smiles', 'state'
]].rename(columns={'moldb_smiles': 'canonical_smiles'})

sweetener_core = Sweetener[[
    'compound_id', 'canonical_smiles', 'logSw'
]]

# Combine cores with priority: BitterDB > FooDB > SweetenerDB
core_df = pd.concat([
    bitter_core,
    foodb_core,
    sweetener_core
], ignore_index=True).drop_duplicates('compound_id')
```

```python
# BitterDB classifications
bitter_taste = citterresourceA.rename(columns={'cID': 'compound_id'})[['compound_id', 'bitterClassification']]

# FooDB flavor mappings
foodb_flavor = pd.merge(
    CompoundFlavor,
    Flavor.rename(columns={'id': 'flavor_id'}),
    on='flavor_id'
)[['compound_id', 'name']].rename(columns={'name': 'taste_class'})

# Sweetener classification
sweetener_taste = Sweetener[['compound_id']]
sweetener_taste['taste_class'] = 'Sweet'

# Combine all taste data
taste_df = pd.concat([
    bitter_taste.rename(columns={'bitterClassification': 'taste_class'}),
    foodb_flavor,
    sweetener_taste
], ignore_index=True)

# Merge with core dataset
merged_df = pd.merge(
    core_df,
    taste_df,
    on='compound_id',
    how='left'
)
```

```python
# BitterDB receptor data
receptor_data = pd.merge(
    ligandReceptorsA.rename(columns={'cID': 'compound_id'}),
    ReceptorSearchA[['rID', 'rName', 'classification']],
    on='rID'
)[['compound_id', 'rName', 'classification', 'EC50microM']]

# Merge receptor interactions
final_df = pd.merge(
    merged_df,
    receptor_data,
    on='compound_id',
    how='left'
)
```

```python
# Handle missing values
final_df['taste_class'] = final_df['taste_class'].fillna('Unknown')
final_df['EC50microM'] = final_df.groupby('taste_class')['EC50microM'].transform(
    lambda x: x.fillna(x.median())
)

# Remove invalid entries
final_df = final_df[~final_df['canonical_smiles'].isna()]

# Save final dataset
final_df.to_csv('Integrated_Taste_Dataset.csv', index=False)
print(f"Final dataset contains {len(final_df)} compounds with {len(final_df.columns)} features")
```
